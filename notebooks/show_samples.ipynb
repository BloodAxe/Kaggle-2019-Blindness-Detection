```python
from pprint import pprint
import cv2
import matplotlib.pyplot as plt
from pytorch_toolbelt.utils.fs import id_from_fname

from retinopathy.dataset import get_aptos2019, get_class_names
tx, vx, ty, vy = get_aptos2019('../data/aptos-2019', 42, 0, 4)
```

```python
def show_samples(x,y, label, num_samples):
    mask = y == label
    x = x[mask][:num_samples]
    y = y[mask][:num_samples]
    class_names = get_class_names()
    for i in range(num_samples):
        image = cv2.imread(x[i])
        image = cv2.cvtColor(image,cv2.COLOR_RGB2BGR)
        
        plt.figure()
        plt.imshow(image)
        plt.title(f'{x[i]} {class_names[y[i]]}')
        plt.axis('off')
        plt.tight_layout()
        plt.show()
        
```
